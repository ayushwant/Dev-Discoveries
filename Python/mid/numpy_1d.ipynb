## What is Numpy?
NumPy is a Python library used for working with arrays, linear algebra, fourier transform, and matrices. NumPy stands for Numerical Python and it is an open source project. The array object in NumPy is called **ndarray**, it provides a lot of supporting functions that make working with ndarray very easy.

Arrays are very frequently used in data science, where speed and resources are very important.

NumPy is usually imported under the np alias.

It's usually fixed in size and each element is of the same type. We can cast a list to a numpy array by first importing `numpy`:


```python
# import numpy library

import numpy as np 
```

```python
a = np.array( [0,1,2,3,4,5] )
a
```

```python
print(type(a))

print(a[0])
print(np.__version__)
print(a.dtype)

a[0] = -1
print(a)
```

```python
#slicing

print(a[1:4]) 

a[1:4] = 100,200,300 # set values of indices [1-4)

print(a)

# We can also define the steps in slicing, like this: [start:end:step].
print(a[0::2])
```

###  Select or Assign Value with List

We can select or assign multiple values of our array by using a list of indices. 


```python
indices = [0,3,4]

arr = a[indices]
print(arr)

a[indices] = -69
print(a)
print(arr)
```

```python
print(a.ndim)
print(a.shape)  # output = (6,). Means 1d array with 6 values
```

### Numpy Statistical Functions


```python
print(a.mean())
print(a.std())
print(a.max())
print(a.sum())
print(a.cumsum())
```

```python
# Plotting functions


import time 
import sys
import numpy as np 

import matplotlib.pyplot as plt


def Plotvec1(u, z, v):
    
    ax = plt.axes() # to generate the full window axes
    ax.arrow(0, 0, *u, head_width=0.05, color='r', head_length=0.1)# Add an arrow to the  U Axes with arrow head width 0.05, color red and arrow head length 0.1
    plt.text(*(u + 0.1), 'u')#Adds the text u to the Axes 
    
    ax.arrow(0, 0, *v, head_width=0.05, color='b', head_length=0.1)# Add an arrow to the  v Axes with arrow head width 0.05, color red and arrow head length 0.1
    plt.text(*(v + 0.1), 'v')#Adds the text v to the Axes 
    
    ax.arrow(0, 0, *z, head_width=0.05, head_length=0.1)
    plt.text(*(z + 0.1), 'z')#Adds the text z to the Axes 
    plt.ylim(-2, 2)#set the ylim to bottom(-2), top(2)
    plt.xlim(-2, 2)#set the xlim to left(-2), right(2)
```

```python
# Plot numpy arrays

Plotvec1(a,b,c)
```

```python
c = np.subtract(a,b)
print(c)
Plotvec1(a,b,c)
```

```python
c = np.multiply(a,b)
print(c)
Plotvec1(a,b,c)
```

```python
a = np.array([1, 2])
b = np.array([2, 1])
c = np.divide(a, b)
print(c)

# c = c.astype(np.int64)
print(c)
Plotvec1(a,b,c)
```

```python
# Dot product
dot = np.dot(a,b)
print(dot)

# maths using scalar to np array
a = a+10
print(a)
a = a-10
print(a)
a = a*10
print(a)
a = a/10
print(a)
```

## Linspace
A useful function for plotting mathematical functions is <code>linspace</code>.   Linspace returns evenly spaced numbers over a specified interval. 

 **numpy.linspace(start, stop, num = int value)**
 
start  :  start of interval range

stop   :  end of interval range

num    :  Number of samples to generate.


```python
a = np.linspace(1,10,num=10)
print(a)

a = np.linspace(1,10,num=19)
print(a)

a = np.linspace(-2, 2, num=9)
print(a)
```

```python

# Make a numpy array within [0, 2π] and 100 elements 

x = np.linspace(0, 2*np.pi, num=100)
# print(x)

# calculate sin
y = np.sin(x)

plt.plot(x,y)

x = np.linspace(0, 2*np.pi, num=5)
y = np.sin(x)
plt.plot(x,y)
```
