### Exercise 1: Pandas: DataFrame and Series 

**Pandas** is a popular library for data analysis built on top of the Python programming language. Pandas generally provide two data structures for manipulating data, They are: 
 
* DataFrame
* Series

A **DataFrame** is a two-dimensional data structure, i.e., data is aligned in a tabular fashion in rows and columns.

* A Pandas DataFrame will be created by loading the datasets from existing storage. 
* Storage can be SQL Database, CSV file, Excel file, etc. 
* It can also be created from the lists, dictionaries, and from a list of dictionaries.

**Series** represents a one-dimensional array of indexed data.
It has two main components :
1. An array of actual data.
2. An associated array of indexes or data labels.

The index is used to access individual data values. You can also get a column of a dataframe as a **Series**. You can think of a Pandas series as a 1-D dataframe. 


A DataFrame is a 2-dimensional data structure that can store data of different types (including characters, integers, floating point values, categorical data and more) in columns. It is similar to a spreadsheet, a SQL table or the data.frame in R.

#### -> Each column in a DataFrame is a Series

### ILOC & LOC

One way to access unique elements is the <code>iloc & loc</code> method, where you can access the 1st row and the 1st column as follows:

```python
# using indices
print(df.iloc[0,0])
```

```python
# using column name => LOC

print( df.loc[0, 'Artist'] ) # give me the 0th artist
print( df.loc[0, 'Genre'] )# give me the 0th genre
```

### Slicing
We can also slice dataframes to get required data.

The indexes of iloc are <code>[inclusive, exclusive)</code>, whereas loc includes both: <code>[inclusive, inclusive]</code>.

```python
df.head()
```

```python
df.iloc[ 0:3, 0:4 ]  # give me indexes [0-3) of the columns [0,4)
```

```python
df.loc[ 0:3, 'Artist':'Length']   # give me indexes [0-3] of the columns [Artist, Length]
```

### Data overview

describe and info are two methods provided by pandas DataFrame for different purposes:

describe: This method is used for generating descriptive statistics of a DataFrame or Series. It gives you a summary of central tendency, dispersion, and shape of the distribution of a dataset, excluding NaN values. By default, it provides the count, mean, standard deviation, minimum, 25th percentile, 50th percentile (median), 75th percentile, and maximum of the columns with numerical data.


info: This method is used to print a concise summary of a DataFrame including the index dtype and column dtypes, non-null values, and memory usage. It's a quick way to understand the structure and data types of your DataFrame.

So, in summary, describe is more about the data content while info is more about the data structure.

```python
# The describe() method provides a quick overview of the numerical data in a DataFrame. 
# The textual data is not considered for the overview. 
df.describe() 
```

### Filtering data in a df


```python
y = df[['Artist','Album','Released', 'Genre']]
y
```

```python
rel_1980 = df[ df['Released'] > 1980 ]
rel_1980
```

```python
# this is a series of boolean values
list_bool = df['Released'] > 1980
list_bool
```

```python
# we can also pass a list of boolean values to filter the data
df[list_bool]
```

```python
# now suppose I'm only interested in these columns, which were released after 1980
y = df[['Artist','Album', 'Genre']] [df['Released'] > 1980]
y
```

```python
# isin also gives a series of boolean values
# keep in mind the dtype

y = df['Released'].isin([1980, 1982])
y
```

```python
y = df[['Artist','Album', 'Genre']] [df['Released'].isin([1980, 1982])]
y
```

```python
y = df[ (df['Released']==1980) | (df['Released']==1982) ]
y
```

```python
# this is again a series of booleans

print(type( (df['Released']==1980) | (df['Released']==1982) ))
print(( (df['Released']==1980) | (df['Released']==1982) ).dtype)
```
